```python
def estandarizar_fecha(fecha):
    try:
        # Convertir la fecha a formato datetime
        fecha_estandar = pd.to_datetime(fecha, errors='coerce')
        # Retornar la fecha en formato "dd/mm/yyyy"
        return fecha_estandar.strftime("%d/%m/%Y")
    except Exception:
        return np.nan  # Retornar NaN si la conversión falla



# Convertir y estandarizar las fechas en la columna 'reservation_status_date'
df['reservation_status_date'] = df['reservation_status_date'].apply(estandarizar_fecha)
```

```python
# Columna IS_CANCELED: Cambiamos los valores True y False a Yes y No

df['is_canceled'] = df['is_canceled'].map({False: 'No', True: 'Yes'})
df.head()
```

```python
# Comprobamos los nulos

for columna in df:
    total_filas = len(df)
    num_nulos = df[columna].isna().sum()
    porcentaje_nulos = (num_nulos / total_filas) * 100
    cuenta_valores = df[columna].value_counts()
    porcentaje_valores = (cuenta_valores / total_filas) * 100
    
    # Mostramos resultados
    print(f"\nColumna: {columna.upper()}")
    print(f"Porcentaje de valores nulos: {porcentaje_nulos:.2f}%")
    print("Porcentaje de cada valor único:")
    print(porcentaje_valores)
```

```python
# Hay varias columnas categóricas que siguen teniendo muchos nulos a pesar de haber eliminado los duplicados.
# Creamos una función para imputar los nulos de estas columnas en función de la distribución de las categorías existentes. De esta manera
# no se distorsionan tanto los resultados.


def impute_categorical(df, column):
    # Calculamos las frecuencias de las categorías existentes
    probabilities = df[column].value_counts(normalize=True)
    # Generamos valores aleatorios basados en la distribución de frecuencias
    impute_values = np.random.choice(probabilities.index, size=df[column].isnull().sum(), p=probabilities.values)
    # Rellenamos los valores nulos con los valores aleatorios generados
    df.loc[df[column].isnull(), column] = impute_values

# Obtenemos solo las columnas categóricas
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Imputamos los valores nulos solo para las columnas categóricas
for col in categorical_columns:
    impute_categorical(df, col)
```

```python
# Comprobamos que, efectivamente, todas las columnas categóricas han sido imputadas correctamente y carecen de nulos

for columna in df:
    total_filas = len(df)
    num_nulos = df[columna].isna().sum()
    porcentaje_nulos = (num_nulos / total_filas) * 100
    cuenta_valores = df[columna].value_counts()
    porcentaje_valores = (cuenta_valores / total_filas) * 100
    
    # Mostramos resultados
    print(f"\nColumna: {columna.upper()}")
    print(f"Porcentaje de valores nulos: {porcentaje_nulos:.2f}%")
    print("Porcentaje de cada valor único:")
    print(porcentaje_valores)
```
